```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


from sklearn import decomposition

classes_cnt = 6

with open('/content/gdrive/My Drive/formulas_test_5_10.txt') as f:
    w = f.read().split('\n')
    inds = [s.count('+') for s in w]
    inds = np.array(inds)
    
with open('/content/gdrive/My Drive/test_5_10_z_8_e_8_h_16.z') as f:
    w = f.read().split('\n')
    ar = [[float(s) for s in v.split()] for v in w[:-1]]
    ar = np.array(ar)

fig = plt.figure(1, figsize=(15, 10))
plt.clf()
ax = Axes3D(fig)

plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(ar)
print(ar.shape)
X = pca.transform(ar)
X1 = X[inds < classes_cnt]
inds = inds[inds < classes_cnt]
print(inds.shape)
print(X.shape)

for name, label in [(str(i), i) for i in range(classes_cnt)]:
    ax.text3D(X1[inds == label, 0].mean(),
              X1[inds == label, 1].mean(),
              X1[inds == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=1, edgecolor='w', facecolor='w'))
    
ax.scatter(X1[:, 0], X1[:, 1], X1[:, 2], c=inds, cmap=plt.cm.nipy_spectral,
           edgecolor='k')
```
