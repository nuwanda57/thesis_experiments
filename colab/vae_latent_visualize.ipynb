```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


from sklearn import decomposition
```

```
def visualize_latent(classes_cnt, test_path, rec_z_path):
    with open(test_path) as f:
        w = f.read().split('\n')
        polynom_degrees = np.array([s.count('+') for s in w])
    
    with open(rec_z_path) as f:
        w = f.read().split('\n')
        zs = np.array([[float(s) for s in v.split()] for v in w[:-1]])

    pca = decomposition.PCA(n_components=3)
    pca.fit(zs)
    zs_transformed = pca.transform(zs)
    zs_transformed = zs_transformed[polynom_degrees < classes_cnt]
    polynom_degrees = polynom_degrees[polynom_degrees < classes_cnt]

    fig = plt.figure(1, figsize=(15, 10))
    # plt.clf()
    ax = Axes3D(fig)

    print(zs_transformed.shape)

    # plt.cla()

    for name, label in [(str(i), i) for i in range(classes_cnt)]:
        ax.text3D(zs_transformed[polynom_degrees == label, 0].mean(),
                  zs_transformed[polynom_degrees == label, 1].mean(),
                  zs_transformed[polynom_degrees == label, 2].mean(), name,
                  horizontalalignment='center',
                  bbox=dict(alpha=1, edgecolor='w', facecolor='w'))
        
    ax.scatter(zs_transformed[:, 0], zs_transformed[:, 1], zs_transformed[:, 2],
               c=polynom_degrees, cmap=plt.cm.nipy_spectral, edgecolor='k')
    ax.legend()
    plt.show()
```
